```python
#basically it is list of dictionary but not exactly a list because it is in weird format 
#it is string of list 
#we have to convert it into list format 
movies.iloc[0].genres
```

```python
#python has function called literal eval which can convert string into a list format

import ast 
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
#here we converted genres into a list format  
movies['genres'].apply(convert)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#converting cast feild into a list 
#so we only need top 3 actors from the list, because on the basis of that audience watches movies
import ast 
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj): #at the begining counter value will be zero
        if counter != 3:
            L.append(i['name']) #will append and increase counter value by one, this loop will continue till counter value become 3
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast']= movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#now there is only one field remaining to be in a right format which is crew 
movies.iloc[0].crew
```

```python
#so in the crew feild we only need a info where job is director 
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
            
        
    return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
#now overview column is also in string format we have to convert it into a list format
movies.iloc[0].overview
```

```python
movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
#now every column is in right format 
movies.head()
```

```python
# we have to do a transformation here now, we have to remove the space between two words 
# if there is space, then it can be problematic for the algorithm 
# example : Sam Worthington and Sam Mendes are two different names but algorithm will interprete them same 
#because the initial word 'Sam' is same between them so algorithm will recommend sam mendes movies also when
# we seach Sam Worthington movies 
#to deal with this problem we have to trasform them into single keyword 
# example : Sam Worthington = SamWorthington
```

```python
# we will use list comprehention
movies['genres'].apply(lambda x:[i.replace(" ","") for i in x ])
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x ])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x ])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x ])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x ])
```

```python
# now we have to concatinate all the columns together into one column which will be tags 
movies['tags'] = movies['overview']+ movies['genres']+ movies['keywords']+ movies['cast']+ movies['crew']
```

```python
# now we wil create a new database which will only havethree  columns 
new_df = movies[['id','title','tags']]
```

```python
# now we have one problem in these 5000 words words like actor actors, agent agents are similar but they 
#here they are adding extra unuseful features in the data 
#there is one technique known as stemming which will convert them into a single word
#for that we will install one liabrary 'nltk'
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
# here we are doing text vectorization we can do it manually but we already have liabrary 'countvectorizer'
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features= 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
#now we have every movie is in vector form 
```

```python
# we can see the 5000 words which are most used 
cv.get_feature_names()
```

```python
#now we have 4806 movies so as 4806 vectors
#We have to calculate distance from each movie to each movie
#we are not going to use Euclidean Distace because it will fail in the high dimentional data 
#instead of that we will use cosine distance which will calculate degree angle
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
#it is showing the similarity of first movie to each movie 
similarity[0]
```

```python
sorted(similarity[0],reverse=True) # obviously we are getting result but we are loosing the indexes that's why we will use enumerate function
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x:x[1])[1:6] # now each and every result has there index
#these are top 5 movies 
```

```python
# now we have to built a function, which will recommend 5 similar movies 
# how it will work?
# first we will find the index of that perticular movie, and will find the similarity
# then we will sort them and will show top 5 movies 
```

```python
def recommend(movie): #created function, recommend 
    movie_index = new_df[new_df['title']== movie].index[0] #it will give us movie_index which we are searching
    distances = similarity[movie_index] # here we are calculating closest vector distancesfrom the movie index 
    movies_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:6]
#     print(movie)
    for i in movies_list:
        
        print(new_df.iloc[i[0]].title)
    return
    
```

```python
recommend('Batman Begins')
```
